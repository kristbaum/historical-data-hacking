# Task 4 – Publikationsstrategien und Forschungsdatenmanagement (FDM)

Auseinandersetzung mit etablierten und innovativen Wegen zur wissenschaftlichen Publikation digitaler Forschungsergebnisse sowie eine fundierte Einführung in die Prinzipien und Praktiken des Forschungsdatenmanagements: Open-Data-Prinzipien, Lizenzierungsfragen, Urheberrechtsaspekte und die Nutzung offener kollaborativer Plattformen für historische Daten (Zenodo, Open Data, Lizenzen, Urheberrecht, Wikisource, OpenHistoricalMap).

## II. Rechtliche und ethische Aspekte
- Open-Data-Prinzpien: Daten müssen vollständig, verfügbar, nutzbar, maschinenlesbar und frei (offene Lizenz) sein.
- [FAIR-Prinzipien](https://de.wikipedia.org/wiki/FAIR-Prinzipien): Daten müssen auffindbar, zugänglich, interoperabel und wiederverwendbar sein.
- Lizensierungsfragen: CC0 und CC BY als Standardlizenzen für Forschungsdaten
- Urheberrecht und Ethik
## III. Wissenschaftliche Publikationsstrategien für digitale Forschungsergebnisse
kollaborative Plattformen für historische Daten
- [Zenodo](https://de.wikipedia.org/wiki/Zenodo): offener Online-Speicherdienst für wissenschaftliche Datensätze (aber auch für wissenschaftsbezogene Publikationen, Berichte, Videos etc.
- [GitHub/GitLab](https://www.heise.de/tipps-tricks/GitHub-vs-GitLab-4597154.html) (in Verbindung mit Zenodo): Repository für Daten und Codes; durch eine Anbindung an Zenodo kann das Repository mit einer DOI versehen und somit zitierfähig gemacht werden.
- [Wikisource](https://de.wikisource.org/wiki/Hauptseite): Sammlung von Texten und Quellen, die entweder urheberrechtsfrei sind oder unter einer freien Lizenz stehen.
- [OpenHistoricalMap](https://www.openhistoricalmap.org/#map=7/47.070/10.865&layers=O): Fokus auf Kartierung historischer Daten
