# Task 2 – Datenanreicherung & Semantische Verknüpfung (Wikidata / GND)

Ziel: Personen- und Objekt-Daten (`people.csv` und `building_openrefine.csv`) mit externen Normdaten abgleichen (Wikidata, GND) und strukturierte Zusatzinformationen (Lebensdaten, alternative Namen, Berufe) übernehmen.

## 1. Wikidata Reconciliation mit Gebäuden (OpenRefine)

1. Importiere `building_openrefine.csv` oder arbeite mit der noch offenen bereinigten Version weiter
2. Wähle im Menü der Spalte "appellation": Reconcile → "Start Reconciling" → mit dem ersten Service; URL: `https://wikidata.reconci.link/en/api` (oder passender anderer Sprach-Endpunkt).
3. Reconcile Spalte `appellation` gegen Wikidata (Type Suggestion "building" / Q41176). Reconciliation mit Typ bringt meistens bessere Ergebnisse, da Resultate, die dem Typ entsprechen, höher gewichtet werden.
4. Reconcile 3 noch nicht reconciliierte Gebäude mit dem korrekten Wikidata-Eintrag. Klicke dazu auf die "judgement" Facet und wähle "none". Und bei der "best candidates score" Facet wähle den Haken bei "blank" ab, um Ergebnisse mit Vorschlägen zu bekommen.

## 2. Wikidata Enrichment (Lebensdaten)

1. Importiere `people.csv`
2. Wähle im Menü der Spalte `wikidata_id` "Reconcile" / "Use values as identifiers"
3. Wähle im Menü der Spalte `wikidata_id` "Edit column" / "Add columns from reconciled values".
4. Gebe in den Suchschlitz folgende Eigenschaften zum Hinzufügen zur Tabelle an:
   - P569 (Geburtsdatum)
   - P570 (Sterbedatum)
   - P19 (Geburtsort; Label)
   - P20 (Sterbeort; Label)
   - P106 (Beruf; Label)
5. Klicke auf OK (Download kann einige Sekunden dauern)

## 3. GND Reconciliation (GND / VIAF)

Direkter GND-Abgleich über den OpenRefine Reconciliation Service bei LobID (ein Service des Hochschulbibliothekszentrums des Landes NRW).
Um Zeit zu sparen: Textfilter im Menü der Spalte "appellation" auswählen und dort nach "Wilhelm" suchen und das Datenset auf Wilhelms begrenzen!
Dann mit dem Knopf "Add Standard Service" diesen Endpunkt hinzufügen: "https://lobid.org/gnd/reconcile"
Im Reconciliation-Fenster "Individualisierte Personen" auswählen.

Optional: VIAF als Reconciliation Service hinzufügen und ausführen: https://refine.codefork.com/reconcile/viaf



## 4. SPARQL Validierung (Wikidata Query Service)

Führe eine Beispiel-Abfrage aus (manuell im WDQS), um z. B. alle gefundenen `wikidata_id` zu verifizieren:

```sparql
SELECT ?person ?personLabel ?birth ?death WHERE {
   VALUES ?person { wd:Q42 wd:Q5582 }  # Beispiel: Ersetze mit echten IDs
   OPTIONAL { ?person wdt:P569 ?birth }
   OPTIONAL { ?person wdt:P570 ?death }
   SERVICE wikibase:label { bd:serviceParam wikibase:language "de,en". }
}
LIMIT 50
```
